#### Question:

Use the `requests` package to:
    
1. Perform a GET request to retrieve information about "how many people are in space right now." 

2. Print the response status code.

3. Parse the JSON into a Python dictionary
    using the following API:
        
4. From the dictionary, print number of astronauts, followed by a list of the names of each astronaut.

Documentation for this API:  
http://open-notify.org/Open-Notify-API/People-In-Space/
    
API endpoint:  
"http://api.open-notify.org/astros.json"

```python
import requests
```

```python
# Answer:

url = 'http://api.open-notify.org/astros.json'

response = requests.get(url)
```

```python
print(response.status_code)
```

```python
import json

people = {}

if response.ok:
    people = json.loads(response.text)
    print(people)
```

```python
print('People in space:', people['number'])

for person in people['people']:
    print(person['name'])
```
