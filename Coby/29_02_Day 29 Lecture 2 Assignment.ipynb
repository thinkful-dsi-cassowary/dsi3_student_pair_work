Write a function to compute entropy given an input of a sequence of probabilities.

```python
#answer below:

from math import log2

def get_entropy(probs):
    entropy = 0
    for prob in probs:
        if prob > 0:
            curr_entropy = (0 - prob) * log2(prob)
        else:
            curr_entropy = 0
        entropy += curr_entropy
    return entropy

probabilities = [(100/500), (300/500), (100/500)]

get_entropy(probabilities)
```

Aggregate the tennis decision table for each value of each column. Start with Outlook below. Compute the weighted mean of the entropy for outlook (the weighted mean of the yes decision and the no decision).

```python
crosstab = pd.crosstab(tennis['Decision'], tennis['Outlook'])
crosstab
```

```python
# answer below:
def get_weighted_entropy(crosstab):
    target = {}
    weighted_total = 0
    for ind in crosstab.index:
        target[ind] = {}
        target[ind]['total'] = tennis['Decision'].value_counts()[ind]
        weighted_total += target[ind]['total']
        probabilities = [crosstab.loc[ind, col]/target[ind]['total'] for col in crosstab.columns]
        target[ind]['entropy'] = get_entropy(probabilities)
    weighted_entropies = [target[key]['entropy'] * (target[key]['total'] / weighted_total) for key in target]

    return np.sum(weighted_entropies)


get_weighted_entropy(crosstab)
```

Compute the weighted mean of the entropy for temperature, humidity and wind as well and decide based on these values which should be the first variable chosen for a split.

```python
# answer below:

for col in ['Temp.', 'Humidity', 'Wind']:
    print('The weighted entropy for {} is {}'.format(col, get_weighted_entropy(pd.crosstab(tennis['Decision'], tennis[col]))))
```
