### Create an interactive histogram to explore the distributions of daily (by date) Quantity, UnitPrice, and Revenue. Include a dropdown widget where you can switch between the three fields.

**Hint:** You will need to create a Date field and then aggregate on it. The Quantity and Revenue fields should be summed and the UnitPrice field should be averaged (mean).

```python
retail['InvoiceDate'] = pd.to_datetime(retail['InvoiceDate'])
retail['Date'] = retail['InvoiceDate'].dt.date
retail
```

```python
daily = retail.groupby('Date', as_index=False).agg({'Quantity': 'sum', 'UnitPrice': 'mean', 'Revenue': 'sum'})
daily
```

```python
def ihistogram(df, field, title=''):
    fig = px.histogram(df, x=field, title=title, template='none').update(layout={'title': {'x': .5}})
    
    fig.update_yaxes(title_text='Number of Records')
    fig.update_traces(marker_color='lightskyblue',
                        marker_line_color='black',
                        marker_line_width=1)
    fig.show()
```

```python
@interact(Data=set(daily.columns.drop('Date')))

def viz(Data):
    ihistogram(daily, field=Data, title='Distribution of Daily {}'.format(Data))
```

### Create an interactive line chart where you can explore Quantity and Revenue by day (date) for each country. Include two dropdown widgets - one to choose between the Quantity and Revenue fields and the other to choose the Country.

```python
daily_country = retail.groupby(['Country', 'Date'], as_index=False).agg(
                                {col: 'sum' for col in ['Quantity', 'Revenue']})
daily_country.head()
```

```python
def ilinechart(df, x, y, groups=None, title=''):
    fig = px.line(df, x=x, y=y, color=groups, title=title,
                 template='none').update(layout={'title': {'x': .5}})
#     for item in range(len(fig.data)):
#         fig.data[item].update(name=fig.data[item]['name'].split('=')[1])
        
    fig.show()
```

```python
@interact(Data=['Quantity', 'Revenue'],
         Country=list(daily_country['Country'].unique()))

def viz(Data, Country):
    this_country = daily_country.loc[daily_country['Country'] == Country]
    ilinechart(this_country, 'Date', Data, title='Daily {} in {}'.format(Data, Country))
```

### Create a bar chart showing the top X products sold (by quantity) in the United Kingdom in a specific month. Use a dropdown box to select the month name and a slider to show the top X products. The range for X should be from 5 to 25. Make sure the bars are sorted in descending order according to their total quantity sold for the month.

Bonus points if you can figure out how to ensure that the month names in the dropdown appear in their correct order (January, February, March, April, etc.).

```python
uk_retail = retail.loc[retail['Country'] == 'United Kingdom']
uk_retail = uk_retail.assign(Month=uk_retail.loc[:,'InvoiceDate'].dt.strftime('%B-%Y'))
monthly_products = uk_retail.groupby(['Month', 'StockCode'], as_index=False).agg({'Quantity': 'sum'})
monthly_products
```

```python
monthrange = list(uk_retail['Month'].unique())
monthrange
```

```python
def ibarchart(df, x, y, order=None, title=''):
    fig = px.bar(df, x=x, y=y, title=title, template='none')
    
    fig.update_traces(marker_color='lightskyblue',
                     marker_line_color='black',
                     marker_line_width=1)
    
    fig.update_layout(xaxis={'categoryorder': 'array',
                            'categoryarray': order})
    
    fig.show()
```

```python
@interact(Month=monthrange, 
         Number_of_Products=(5,25,1))

def viz(Month, Number_of_Products):
    
    this_month = monthly_products.loc[monthly_products['Month'] == Month]
    top_products = this_month.sort_values(by='Quantity', ascending=False).head(Number_of_Products)
    top_products['StockCode'] = top_products['StockCode'].apply(lambda x: f'SC:{x}')
    ibarchart(top_products, 'StockCode', 'Quantity', title='Top {} Products for {}'.format(Number_of_Products, Month))
```
